```python
import pandas as pd
```

```python
btc_data = pd.read_csv("BTC-USD.csv")
```

```python
btc_data 
```

```python
print("First 5 Rows:")
print(btc_data.head())
```

```python
print("\nData Information:")
print(btc_data.info())
```

```python
print("\nSelected Columns (Date and Close):")
print(btc_data[['Date', 'Close']].head())
```

```python
filtered_data = btc_data[(btc_data['Date'] >= '2020-01-01') & (btc_data['Date'] <= '2020-12-31')]
print("\nFiltered Data for 2020:")
print(filtered_data.head())
```

```python
print("\nAverage Closing Price:", btc_data['Close'].mean())
print("Total Trading Volume:", btc_data['Volume'].sum())
```

```python
print("\nDescriptive Statistics:")
print(btc_data.describe())
```

```python
btc_data['Daily Range'] = btc_data['High'] - btc_data['Low']
btc_data['Daily Return (%)'] = ((btc_data['Close'] - btc_data['Open']) / btc_data['Open']) * 100
print("\nNew Columns Added:")
print(btc_data[['Date', 'Daily Range', 'Daily Return (%)']].head())
```

```python
btc_data_dropped = btc_data.drop(['Adj Close'], axis=1)
print("\nDropped 'Adj Close' Column:")
print(btc_data_dropped.head())
```

```python
sorted_data = btc_data.sort_values(by='Volume', ascending=False)
print("\nTop 5 Days by Volume:")
print(sorted_data[['Date', 'Volume']].head())
```

```python
btc_data.at[10, 'Open'] = None  
btc_data['Open'] = btc_data['Open'].fillna(btc_data['Open'].mean())
print("\nAfter Handling Missing Data:")
print(btc_data.iloc[8:12])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.plot(pd.to_datetime(btc_data['Date']), btc_data['Close'], label='Close', color='blue')
plt.title('Bitcoin Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(btc_data['Close'], bins=20, color='green', alpha=0.7)
plt.title('Distribution of Closing Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
top_10_volumes = btc_data.nlargest(10, 'Volume')
plt.figure(figsize=(8, 5))
plt.bar(top_10_volumes['Date'], top_10_volumes['Volume'], color='orange')
plt.title('Top 10 Trading Volumes')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(btc_data['High'], btc_data['Low'], color='purple', alpha=0.5)
plt.title('High vs. Low Prices')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.show()
```

```python
price_ranges = [btc_data[btc_data['Close'] <= 10000].shape[0],
                btc_data[(btc_data['Close'] > 10000) & (btc_data['Close'] <= 20000)].shape[0],
                btc_data[btc_data['Close'] > 20000].shape[0]]
labels = ['<= 10k', '10k-20k', '> 20k']
plt.figure(figsize=(6, 6))
plt.pie(price_ranges, labels=labels, autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green'])
plt.title('Distribution of Closing Prices by Range')
plt.show()
```
